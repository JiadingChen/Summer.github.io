# RiceQuant is based on virtual env for windows
### Windows10, Aqalpha4.0.1a, Python 3.8 
1. conda create -n rice python=3.8
2. activate rice 
3. pip install rqalpha
4. rqalpha version 查看是否成功
5. rqalpha download-bundle 注入数据，bundle 默认存放在 ~/.rqalpha 下，您也可以指定 bundle 的存放位置
6. rqalpha update-bundle 更新回测数据
7. rqalpha generate-config 获取配置文件
8. conda install -c conda-forge tensorflow
9. pip install ipykernel
10. python -m ipykernel install --user --name=rice
11. Dowload ta_lib from https://www.lfd.uci.edu/~gohlke/pythonlibs/#ta-lib
12. pip install TA_Lib-0.4.18-cp37-cp37m-win_amd64.whl
13. pip install h5py --upgrade --no-dependencies --force
13. deactivate
